```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.3
n = 1000
tosses = []
turn = []
ns = []
ys = []

for i in range (1, n, 1):
    toss = np.random.uniform(low=0, high=1, size=1)
    if toss[0] > p: 
        tosses.append(0)
    else:
        tosses.append(1)
    turn.append(tosses.count(1))
    ns.append(i)
    ys.append(tosses.count(1)/i)

plt.plot(ns, ys)
```
